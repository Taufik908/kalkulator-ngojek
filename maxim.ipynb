# KALKULATOR PENDAPATAN NGOJEK

```python
saldo = []
saldo_dipotong = 0
pendapatan_tunai = []
komisi_saldo = []
total_tunai = 0
total_komisi = 0
keuntungan = 0

while True:
    # Menu ditampilkan sekali di awal
    print("\nMenu:")
    print("1. Jumlah saldo")
    print("2. Pendapatan Tunai")
    print("3. Komisi")
    print("4. Hasil")
    print("5. Keuntungan")
    
    pilihan = int(input("Pilihan Anda: "))

    if pilihan == 1:
        saldo = int(input("Total saldo: "))
        if saldo < 0:
            print("Saldo Tidak boleh negatif")
        else:
            print(f"Total saldo: {saldo}")

        lanjut = input("\nKembali ke menu? (y/n): ")
        if lanjut.lower() != 'y':
            break  # Keluar dari program
    
    elif pilihan == 2:
        while True:  # Loop untuk input pendapatan
            pendapatan_tunai = int(input("Pendapatan Tunai: [0 untuk berhenti] "))
            
            if pendapatan_tunai == 0:
                print(f"Total pendapatan: {total_tunai}")
                print("Selesai")
                break  # Keluar dari loop input pendapatan
            elif pendapatan_tunai < 0:
                print("Nilai tidak boleh negatif")
            else:
                total_tunai += pendapatan_tunai
                print(f"Total saat ini: {total_tunai}")

        # Tanya apakah ingin kembali ke menu
        lanjut = input("\nKembali ke menu? (y/n): ")
        if lanjut.lower() != 'y':
            break  # Keluar dari program

    elif pilihan == 3:
        while True:  # Loop untuk input komisi
            komisi_saldo = int(input("Jumlah komisi: [0 untuk berhenti]"))
            
            if komisi_saldo == 0:
                print(f"Total komisi: {total_komisi}")
                print("Selesai")
                break
            elif komisi_saldo < 0:
                print("Nilai tidak boleh negatif")
            else:
                total_komisi += komisi_saldo
                print(f"Total saat ini: {total_komisi}")

        # Tanya apakah ingin kembali ke menu
        lanjut = input("\nKembali ke menu? (y/n): ")
        if lanjut.lower() != 'y':
            break  # Keluar dari program
    
    
    elif pilihan == 4:
        print(f"Saldo: {saldo}")
        print(f"Pendapatan: {total_tunai}")
        print(f"Komisi: {total_komisi}")

        # Tanya apakah ingin kembali ke menu
        lanjut = input("\nKembali ke menu? (y/n): ")
        if lanjut.lower() != 'y':
            break  # Keluar dari program
    
    elif pilihan == 5:
        keuntungan = total_tunai - total_komisi
        saldo_dipotong = saldo - total_komisi
        print(f"Saldo awal: {saldo}")
        print(f"Saldo akhir: {saldo_dipotong}")
        print(f"Modal(komisi): {total_komisi}")
        print(f"Pendapatan: {total_tunai}")
        print(f"Keuntungan: {keuntungan}")

        # Tanya apakah ingin kembali ke menu
        lanjut = input("\nKembali ke menu? (y/n): ")
        if lanjut.lower() != 'y':
            break  # Keluar dari program

    else:
        print("Pilihan tidak valid.")

print("Program selesai")

```
